Os dados do discurso foram retirados das transcrições do site da Câmara dos Deputados e tratados pelo Marco Zero. Podem conter pequenos erros de digitação.

```python
import re
import string
frequency = {}
document_text = open('003.temer/votacaotemer.txt', 'r')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
 
for words in frequency_list:
    print frequency[words], words
    
```
